```python
import os
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
from IPython.display import Image
```

```python
IMAGE_PATH = '/content/drive/MyDrive/tata.png'
```

```python
img = cv2.imread(IMAGE_PATH)
plt.imshow(img)
plt.show()
```

```python
reader = easyocr.Reader(['en'])
result = reader.readtext(IMAGE_PATH,paragraph="False")
result
```

```python
cd = result[-1][0]
```

```python
x_min , y_min = [min(idx) for idx in zip(*cd)]
```

```python
x_max , y_max = [max(idx) for idx in zip(*cd)]
```

```python
img = cv2.imread(IMAGE_PATH)
img = cv2.rectangle(img,(x_min , y_min),(x_max , y_max),(0,0,255),2)
plt.imshow(img)
plt.show()
```
